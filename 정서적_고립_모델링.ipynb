## 목적변수 Y 전처리

```python
import pandas as pd
import numpy as np
pd.options.display.max_columns = None
data1 = pd.read_csv("/2021_한국행정연구원_사회통합실태조사_데이터.csv", encoding='euc-kr', index_col = 0)
data1.head()
```

```python
ind = data1[data1["b31"]!= "19~29세"].index
data1.drop(ind, inplace = True)
```

```python
data1.reset_index(drop = True, inplace = True)
```

#### (1) 1-4 점수화

```python
data1["q32_1_f"] = data1["q32_1"].replace('매우 그렇다',4)
data1["q32_1_f"] = data1["q32_1_f"].replace('약간 그렇다',3)
data1["q32_1_f"] = data1["q32_1_f"].replace('별로 그렇지 않다',2)
data1["q32_1_f"] = data1["q32_1_f"].replace('전혀 그렇지 않다',1)
```

```python
data1["q32_2_f"] = data1["q32_2"].replace('매우 그렇다',4)
data1["q32_2_f"] = data1["q32_2_f"].replace('약간 그렇다',3)
data1["q32_2_f"] = data1["q32_2_f"].replace('별로 그렇지 않다',2)
data1["q32_2_f"] = data1["q32_2_f"].replace('전혀 그렇지 않다',1)
```

```python
data1["q32_3_f"] = data1["q32_3"].replace('매우 그렇다',4)
data1["q32_3_f"] = data1["q32_3_f"].replace('약간 그렇다',3)
data1["q32_3_f"] = data1["q32_3_f"].replace('별로 그렇지 않다',2)
data1["q32_3_f"] = data1["q32_3_f"].replace('전혀 그렇지 않다',1)
```

```python
data1['감정적 고립'] = data1['q32_1_f'] + data1['q32_2_f'] + data1['q32_3_f']
```

```python
data1['감정적 고립'].value_counts() # 4~12점
```

#### (2) MinMax Scale

```python
mn=data1['감정적 고립'].min()
mx=data1['감정적 고립'].max()
data1['감정적 고립mm'] = data1['감정적 고립'].apply(lambda x:(x-mn)/(mx-mn))
```

```python
data1['감정적 고립mm'].value_counts()
```

### (4) DNN

```python
dataset=df.drop(['id'],axis=1)
dataset
```

```python
train_dataset = dataset.sample(frac=0.8,random_state=0)
test_dataset = dataset.drop(train_dataset.index)
```

```python
train_stats = train_dataset.describe()
train_stats.pop("감정적 고립mm")
train_stats = train_stats.transpose()
train_stats
```

```python
train_labels = train_dataset.pop('감정적 고립mm')
test_labels = test_dataset.pop('감정적 고립mm')
```

```python
def norm(x):
  return (x - train_stats['mean']) / train_stats['std']
normed_train_data = norm(train_dataset)
normed_test_data = norm(test_dataset)
```

```python
def build_model():
  model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=[len(train_dataset.keys())]),
    layers.Dense(64, activation='relu'),
    layers.Dense(1)
  ])

  optimizer = tf.keras.optimizers.RMSprop(0.001)

  model.compile(loss='mse',
                optimizer=optimizer,
                metrics=['mae', 'mse'])
  return model
```

```python
model = build_model()
```

```python
model.summary()
```

```python
# 에포크가 끝날 때마다 점(.)을 출력해 훈련 진행 과정을 표시
class PrintDot(keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs):
    if epoch % 100 == 0: print('')
    print('.', end='')

EPOCHS = 1000

history = model.fit(
  normed_train_data, train_labels,
  epochs=EPOCHS, validation_split = 0.1, verbose=0,
  callbacks=[PrintDot()])
```

```python
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail()
```

```python
def plot_history(history):
  hist = pd.DataFrame(history.history)
  hist['epoch'] = history.epoch

  plt.figure(figsize=(8,12))

  plt.subplot(2,1,1)
  plt.xlabel('Epoch')
  plt.ylabel('Mean Abs Error')
  plt.plot(hist['epoch'], hist['mae'],
           label='Train Error')
  plt.plot(hist['epoch'], hist['val_mae'],
           label = 'Val Error')
  plt.ylim([0,0.2])
  plt.legend()

  plt.subplot(2,1,2)
  plt.xlabel('Epoch')
  plt.ylabel('Mean Square Error')
  plt.plot(hist['epoch'], hist['mse'],
           label='Train Error')
  plt.plot(hist['epoch'], hist['val_mse'],
           label = 'Val Error')
  plt.ylim([0,0.1])
  plt.legend()
  plt.show()

plot_history(history)
```
